Создайте функцию , которая будет применена к столбцу рейтинг методом `apply()`.
Функция должна принимать строку с рейтингом и возвращать:
 - High rating, если рейтинг больше или равен 4.5
 - Middle rating, если рейтинг между 3.8 и 4.5
 - Low rating, если рейтинг меньше 3.8 
Результат работы функции запишите в новый столбец с названием `categorical_rating`

```python
# Создание функции для категоризации рейтинга
def categorize_rating(rating):
    if rating >= 4.5:
        return 'High rating'
    elif rating >= 3.8:
        return 'Middle rating'
    else:
        return 'Low rating'

# Применение функции к столбцу рейтинга
df['categorical_rating'] = df['rating'].apply(categorize_rating)

# Вывод результата
df.head(5)
```

Выведите кол-во уникальных значений в столбцах отсортированных по убыванию на экран:

```python
print(df.nunique().sort_values(ascending=False))
```

Создайте функцию , которая будет применена к столбцу `last updated` и возвращать из него число месяца, если число месяца вернуть из строки невозможно, то верните вместо этого стркоу с названием `miss_date`.\
Запишите полученный столбец в датафрейм , название столбцу дайте `day_of_update`.

Если вы столкнётесь с некоторой ошибкой , которую вам выдаст Pandas, то изучите конструкцию Python `try` \ `except`.\
Вот псевдокод , который поможет лучше разобраться с этой конструкцией:

```python
def get_month(date_str):
    try:
        date = pd.to_datetime(date_str)
        month = date.month
        return month
    except:
        return 'miss_date'
```

```python
# Применяем функцию к столбцу last updated и создаем новый столбец day_of_update
df['day_of_update'] = df['last_updated'].apply(get_month)
#Выводим 5 случайных строк
df.sample(5)
```
