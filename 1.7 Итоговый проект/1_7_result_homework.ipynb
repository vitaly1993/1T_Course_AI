# Итоговый проект.

Анализ личности клиента — это подробный анализ клиентов компании. Он помогает компании лучше понять своих клиентов и облегчает изменение продукции в соответствии со специфическими потребностями, поведением различных типов клиентов.

Анализ личности клиента помогает компании модифицировать свой продукт в соответствии с целевыми потребителями из различных сегментов потребителей. Например, вместо того, чтобы тратить деньги на маркетинг нового продукта для каждого клиента в базе данных компании, компания может проанализировать, какой сегмент клиентов с наибольшей вероятностью купит продукт, и затем продвигать продукт только для этого сегмента.

## Визуализация

Думаю, что пора отвечать на вопросы заказчика

1. Сколько в среднем клиенты тратят на каждый тип товара? На какой товар надо сделать акцент?

```python
# Определеяем средние значения по всем категорям товаров
mean_values = df[['mntwines', 'mntfruits', 'mntmeatproducts', 'mntfishproducts', 'mntsweetproducts', 'mntgoldprods']].mean()
# Строим график
fig = px.bar(mean_values, text_auto=True)
fig.update_layout(title='Среднее значение затрат по категориям товаров',
                  title_x=0.5,
                  xaxis_title='Категории товаров',
                  yaxis_title='Среднее значение затрат',
                  showlegend=False)
```

Больше всего клиенты трятят на вино, затем на мясо. Акцент на мой взгляд следует сделать на фрукты и сладости, может быть расширить ассортимент этих товаров. 

2. Какой средний возраст покупателя? Зависит ли возраст клиента от его покупок во всех категориях?

```python
fig = px.histogram(df, x='year_birth', y=['mntwines', 'mntfruits', 'mntmeatproducts', 'mntfishproducts', 'mntsweetproducts', 'mntgoldprods'])
fig.update_layout(title='Затраты по категориям товаров в зависимости от возраста',
                  title_x=0.5,
                  xaxis_title='Год рождения клиентов',
                  yaxis_title='Затраты по категориям товаров')
```

- Средний возраст покупателя 54 года
- Покупки клиента во всех категориях зависят от его возраста и подчиняются нормальному распределению. Чем ближе возраст клиента к среднему значению тем больше товаров во всех категориях он покупает.


3. Тратит ли клиент больше денег, если у него много детей и иждивенцев?

```python
fig = px.scatter(df, x = df[['kidhome','teenhome']].sum(axis=1), y=df[['mntwines', 'mntfruits', 'mntmeatproducts', 'mntfishproducts', 'mntsweetproducts', 'mntgoldprods']].sum(axis=1))
fig.update_layout(title='Траты клиентов в зависимости от наличия иждевенцев',
                  title_x=0.5,
                  xaxis_title='Количество иждевенцев',
                  xaxis=dict(tickmode='linear', tick0=0, dtick=1),
                  yaxis_title='Траты клента',
                  showlegend=False)
fig.show()
```

Нет, наоборот тратит люди имеющие иждевенцев тратят меньше

4. Зависят ли покупки от образования? Что предпочитают люди чаще покупать в зависимости от образования?

```python
fig = px.histogram(df, x='education', y = ['mntwines', 'mntfruits', 'mntmeatproducts', 'mntfishproducts', 'mntsweetproducts', 'mntgoldprods'])
fig.update_layout(title='Зависимость покупок от образования',
                  title_x=0.5,
                  xaxis_title='Образование',
                  yaxis_title='Траты клента')
fig.show()
```

Явного различия в категориях покупок при различных уровнях образования нет. Клиенты с образованием Graduation покупают больше товаров. 

5. Женатые люди тратят больше или меньше денег на покупку золота?

```python
fig = px.histogram(df, x='marital_status', y = 'mntgoldprods')
fig.update_layout(title='Зависимость покупки золота от семейного положения',
                  title_x=0.5,
                  xaxis_title='Семейное положение',
                  yaxis_title='Траты на золото')
fig.show()
```

Женатые тратят на покупку золота больше денег

6. Люди, которые покупают товар на сайте чаще берут что-то со скидкой?

```python
discount_web_yes = df.loc[(df['numwebpurchases']>0) & (df['numdealspurchases']>0)].shape[0]
discount_web_no = df.loc[(df['numwebpurchases']>0) & (df['numdealspurchases']==0)].shape[0]
print(f'Количество клиентов купивших товары через web сайт по скидке - {discount_web_yes}')
print(f'Количество клиентов купивших товары через web без скидки - {discount_web_no}')

fig = px.bar(df, x='numwebpurchases', y='numdealspurchases')
fig.update_layout(title='Зависимость покупок товаров на сайте со скидками',
                  title_x=0.5,
                  xaxis_title='Количество покупок на сайте',
                  yaxis_title='Количество товаров со скидкой')
fig.show()
```

98,2% клиентов купивших товары на web сайте сделали это по скидке

7. Зависит ли жалоба клиента от его трат?

```python
# Добавляем столбец с суммой товаров по всем категориям
df['sum_mnt'] = df[['mntwines', 'mntfruits', 'mntmeatproducts', 'mntfishproducts', 'mntsweetproducts', 'mntgoldprods']].sum(axis=1)
# Строим матрицу корреляции 
sns.heatmap(df[['sum_mnt', 'complain']].corr(), annot=True)
plt.title('Матрица корреляции между жалобой и суммой покупки')
```

Зависимости между жалобами и тратами клиентов нет

8. Люди, принявшие предложение о покупке лишь в последней компании тратят больше, чем люди, принявшие сразу предложение первой компании?

```python
#Выделяем клиентов принявших предложение 1 компании
company_1 = df.loc[df['acceptedcmp1']==1, ['acceptedcmp1', 'sum_mnt']]
#Выделяем клиентов принявших предложение в последней компании
response = df.loc[df['response']==1, ['response', 'sum_mnt']]
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20, 5))
sns.boxplot(data=company_1, x='sum_mnt', ax=axs[0])
axs[0].set_xlabel('Сумма затрат клиентов принявших предложение в 1-й компании')
sns.boxplot(data=response, x='sum_mnt', ax=axs[1])
axs[1].set_xlabel('Сумма затрат клиентов принявших предложение последней компании')
```

Люди, принявшие предложение о покупке лишь в последней компании тратят меньше чем люди, принявшие сразу предложение первой компании. Это показывает медиана общей суммы затрат.
