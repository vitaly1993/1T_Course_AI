1. Написать программу, которая будет выводить на экран числа Фибоначчи до заданного числа.

```python
n = int(input("Введите число, до которого нужно вывести последовательность Фибоначчи: "))

# Инициализация первых двух чисел Фибоначчи
a, b = 0, 1

# Вывод первых n чисел Фибоначчи, не превышая n
if n >= 0:
    print(a)
    if n >= 1:
        print(b)
    c = a + b
    while c <= n:
        print(c)
        a, b = b, c
        c = a + b
```

2. Напишите программу, которая на вход принимает список чисел и возвращает только уникальные значения из этого списка.

```python
# Вариант 1 (список без повторов)
my_list = [1, 2, 3, 2, 4, 1, 5, 3, 'hello', 'hello', 'bye']
unique_list = list(set(my_list))
print(unique_list)

# Вариант 2 (список с элементами у которых нет повторов)
my_list = [1, 2, 3, 2, 4, 1, 5, 3, 'hello', 'hello', 'bye']
unique_list = []
for element in my_list:
  if element not in unique_list:
    unique_list.append(element)
  else:
    unique_list.remove(element)
print(unique_list)
```

3. Напишите программу, которая переводит температуру в градусах Фаренгейта в градусы Цельсия.

```python
def fahrenheit_to_celsius(fahrenheit):
    celsius = (fahrenheit - 32) * 5/9
    return round(celsius, 2)

while True:
    try:
        temperature_fahrenheit = float(input("Введите температуру в градусах Фаренгейта: "))
        temperature_celsius = fahrenheit_to_celsius(temperature_fahrenheit)
        print(f"{temperature_fahrenheit} градусов Фаренгейта соответствуют {temperature_celsius} градусам Цельсия")
        break
    except ValueError:
        print("Ошибка: введите число")
```

4. Напишите программу, которая проверяет, является ли строка палиндромом без использования стандартных методов.

```python
# Вариант без использования методов python (P.S. использовались только функции)
def is_palindrome(s:str)->bool:
    chars_to_remove = [' ', ',', '.', '!', '?', ';', ':', 'ь', 'ъ', "'", '"', '-']
    new_s = ''
    for char in s:
        if char not in chars_to_remove:
          if 'А' <= char <= 'Я' or 'A' <= char <= 'Z':
            new_s += chr(ord(char) + 32)
          else:
            new_s += char
    reversed_s = new_s[::-1]
    return new_s == reversed_s 
        
s = "а роза упала на лапу Азора"
#s = "Was it a cat I saw?"
if is_palindrome(s):
    print(f"{s} - это палиндром")
else:
    print(f"{s} - это не палиндром")
```
