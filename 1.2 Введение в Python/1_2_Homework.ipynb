2. Напишите функцию, которая будет принимать два числа и считать сумму квадратов этих двух чисел, запишите своё вычисление в переменную **squared**. (Подсказка: Помните, чтобы записать результат функции в переменную, она должна этот результат внутри себя возвращать!).

```python
def sum_square(ch_one:float, ch_two:float) -> float:
  squared = ch_one**2 + ch_two**2
  return squared

# Проверка функции sum_square
print(sum_square(2, 4))
```

3. Напишите функцию, которая будет принимать на вход список  **["male","male","female","male","male","female","female"]** и возвращать его в обратном порядке. Запишите список в переменную **new_list**. Подсказка: Вам поможет обратная индексация к спискам ! **[::-1]**. 

```python
list_gender = ["male","male","female","male","male","female","female"]

def reverse_list_gender(list_gender:list) -> list:
  new_list = list_gender[::-1]
  return new_list

# Проверка функции reverse_list_gender
print(reverse_list_gender(list_gender))
```

4. Напишите функцию, которая будет принимать на вход список (new_list) и идти по нему в цикле. На выходе функция должна вывести на экран: Кол-во мужчин 4, Кол-во женщин 3.

```python
# Вариант 1
def count_gender_badly(list_gender:list):
    male_count = 0
    female_count = 0
    for gender in list_gender:
        if gender == "male":
            male_count += 1
        elif gender == "female":
            female_count += 1
    print("Кол-во мужчин", male_count)
    print("Кол-во женщин", female_count)
    
# Вариант 2
def count_gender_good(list_gender:list):
  male_count = sum([1 for gender in list_gender if gender == "male"])
  female_count = sum([1 for gender in list_gender if gender == "female"])
  print(f"Кол-во мужчин: {male_count}")
  print(f"Кол-во женщин: {female_count}")

# Проверка функции count_gender_good
count_gender_good(reverse_list_gender(list_gender))
# Проверка функции count_gender_badly
count_gender_badly(reverse_list_gender(list_gender))
  
```

5. Напишите функцию, которая будет принимать строку **(Например: "female")**, а возвращать словарь где ключи — это символы строки, а значения кол-во символов в строке. 

```python
# Вариант 1
def count_chars_badly(string: str) -> dict:
    char_count = {}
    for char in string:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    return char_count


# Вариант 2
def count_chars_good(string: str) -> dict:
    char_count = {char: string.count(char) for char in string}
    return char_count


my_string = "female"
# Проверка функции count_chars_good
print(count_chars_good(my_string))
# Проверка функции count_chars_badly
print(count_chars_badly(my_string))

```

6. Воспользуйтесь функцией **range** в Python и выведите на экран сумму квадратов **от 1 до 100 (включительно)**. 

```python
# Вариант 1
print(sum([i**2 for i in range(101)]))

# Вариант 2
total = 0
for i in range(1, 101):
  total += i**2
print(total)


```
